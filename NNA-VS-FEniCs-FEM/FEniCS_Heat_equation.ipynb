## **Heat equation in 1D**

$$\frac{\partial u }{\partial t}=\frac{\partial^2 u}{\partial x^2},\quad in\quad \Omega 𝗫(0,T]= \Omega_T$$$$, \quad u=u_D \quad on \quad ∂Ω𝗫(0,T]$$$$u=u_0 \quad in\quad t=0$$


```python
from __future__ import print_function
from fenics import *
import numpy as np

start_time = time.time()

T = 2.0            # tiempo final
num_steps = 10     # número de pasos de tiempo
dt = T / num_steps # tamaño del paso de tiempo
alpha = 3          # parámetro alpha
beta = 1.2         # parámetro beta

# Definir la malla y el espacio de funciones
nx = 100
mesh = IntervalMesh(nx, 0, 1)
V = FunctionSpace(mesh, 'P', 1)

# Definir la condición de contorno
u_D = Expression('1 + x[0]*x[0] + alpha*t', degree=2, alpha=alpha, t=0)

def boundary(x, on_boundary):
    return on_boundary

bc = DirichletBC(V, u_D, boundary)

# Definir el valor inicial
u_n = interpolate(u_D, V)

# Definir el problema variacional
u = TrialFunction(V)
v = TestFunction(V)
f = Constant(beta - 2)

F = u*v*dx + dt*dot(grad(u), grad(v))*dx - (u_n + dt*f)*v*dx
a, L = lhs(F), rhs(F)

# Iteración en el tiempo
u = Function(V)
t = 0
for n in range(num_steps):

    # Actualizar el tiempo actual
    t += dt
    u_D.t = t

    # Resolver
    solve(a == L, u, bc)

    # Calcular y mostrar el error
    u_e = interpolate(u_D, V)
    error = np.abs(u_e.vector()[:] - u.vector()[:]).max()
    print('t = %.2f: error = %.3g' % (t, error))

    # Actualizar la solución previa
    u_n.assign(u)

# Mostrar la solución
plot(u)


end_time = time.time()
time1 = end_time - start_time
Time_execution.append(time1)
print("Tiempo de ejecución:", time1, "segundos")

# Uso de la memoria
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"Uso de memoria: {mem_info.rss / (1024 ** 2):.2f} MB")
```

## Heat equation in 2D

# 2D espaciales y 1D temporal
$$\frac{\partial u }{\partial t}=(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}),\quad in\quad \Omega 𝗫(0,T]= \Omega_T$$$$, \quad u=u_D \quad on \quad ∂Ω𝗫(0,T]$$$$u=u_0 \quad in\quad t=0$$

```python
from __future__ import print_function
from fenics import *
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline

start_time = time.time()
T = 2.0            # final time
num_steps = 10     # number of time steps
dt = T / num_steps # time step size
alpha = 3          # parameter alpha
beta = 1.2         # parameter beta

# Create mesh and define function space
nx = ny = 8
mesh = UnitSquareMesh(nx, ny)
V = FunctionSpace(mesh, 'P', 1)

# Define boundary condition
u_D = Expression('1 + x[0]*x[0] + alpha*x[1]*x[1] + beta*t',
                 degree=2, alpha=alpha, beta=beta, t=0)

def boundary(x, on_boundary):
    return on_boundary

bc = DirichletBC(V, u_D, boundary)

# Define initial value
u_n = interpolate(u_D, V)
#u_n = project(u_D, V)

# Define variational problem
u = TrialFunction(V)
v = TestFunction(V)
f = Constant(beta - 2 - 2*alpha)

F = u*v*dx + dt*dot(grad(u), grad(v))*dx - (u_n + dt*f)*v*dx
a, L = lhs(F), rhs(F)

# Time-stepping
u = Function(V)
t = 0
for n in range(num_steps):

    # Update current time
    t += dt
    u_D.t = t

    # Compute solution
    solve(a == L, u, bc)

    # Plot solution
    plot(u)

    # Compute error at vertices
    u_e = interpolate(u_D, V)
    error = np.abs(u_e.vector().get_local() - u.vector().get_local()).max()
    print('t = %.2f: error = %.3g' % (t, error))

    # Update previous solution
    u_n.assign(u)

# Hold plot
plt.show()

end_time = time.time()
time2 = end_time - start_time
Time_execution.append(time2)
print("Tiempo de ejecución:", time2, "segundos")

# Uso de la memoria
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"Uso de memoria: {mem_info.rss / (1024 ** 2):.2f} MB")
```

## **Heat equation in 3D**

# 3D espaciales y 1D temporal
$$\frac{\partial u }{\partial t}=(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}+\frac{\partial^2 u}{\partial z^2}),\quad in\quad \Omega 𝗫(0,T]= \Omega_T$$$$, \quad u=u_D \quad on \quad ∂Ω𝗫(0,T]$$$$u=u_0 \quad in\quad t=0$$

```python
from __future__ import print_function
from fenics import *
import numpy as np
import time
start_time = time.time()
T = 2.0            # Tiempo final
num_steps = 10     # Número de pasos de tiempo
dt = T / num_steps # Tamaño del paso de tiempo
alpha = 3          # Parámetro alpha
beta = 1.2         # Parámetro beta

# Creamos la malla y definimos el espacio de funciones
nx = ny = nz = 9
mesh = UnitCubeMesh(nx, ny, nz)
V = FunctionSpace(mesh, 'P', 1)

# Definimos la condición de contorno
u_D = Expression('1 + x[0]*x[0] + alpha*x[1]*x[1] + alpha*x[2]*x[2] + beta*t',
                 degree=2, alpha=alpha, beta=beta, t=0)

def boundary(x, on_boundary):
    return on_boundary

bc = DirichletBC(V, u_D, boundary)

# Definimos el valor inicial
u_n = interpolate(u_D, V)

# Definimos el problema variacional
u = TrialFunction(V)
v = TestFunction(V)
f = Constant(beta - 2 - 2*alpha)

F = u*v*dx + dt*dot(grad(u), grad(v))*dx - (u_n + dt*f)*v*dx
a, L = lhs(F), rhs(F)

# Iteración en el tiempo
u = Function(V)
t = 0
for n in range(num_steps):

    # Actualizamos el tiempo actual
    t += dt
    u_D.t = t

    # Calculamos la solución
    solve(a == L, u, bc)


    # Calculamos el error en los vértices
    u_e = interpolate(u_D, V)
    error = np.abs(u_e.vector().get_local() - u.vector().get_local()).max()
    print('t = %.2f: error = %.3g' % (t, error))

    # Actualizamos la solución previa
    u_n.assign(u)

# Graficamos la solución
from matplotlib import pyplot as plt

# Creamos la figura
fig = plt.figure()

# Obtenemos la malla y los valores de la solución
mesh_coordinates = mesh.coordinates()
u_values = u.compute_vertex_values(mesh)

# Graficamos la solución
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mesh_coordinates[:, 0], mesh_coordinates[:, 1], mesh_coordinates[:, 2], c=u_values, cmap='viridis')

# Configuramos las etiquetas y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Solución de la ecuacion de calor en 3D')

# Mostramos la gráfica
plt.show()

# Imprimimos la solución analítica para el tiempo final
u_D.t = T
u_exact = interpolate(u_D, V)
u_exact_array = u_exact.compute_vertex_values(mesh)
print("Solución analítica en los vértices para t = %.2f:" % T)
print(u_exact_array)


end_time = time.time()
time3 = end_time - start_time
Time_execution.append(time3)
print("Tiempo de ejecución:", time3, "segundos")

# Uso de la memoria
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"Uso de memoria: {mem_info.rss / (1024 ** 2):.2f} MB")
```

# **Ecuación de calor para altas dimensiones** 🔥


Se definen primeramente las funciones que nos ayudaran a hacer los productos tensoriales


```python
#Import modules
from __future__ import print_function
import dolfin as dl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Some auxillary functions defined first
def cartesian_product(array1,array2):
    #gives cartesian product of 2 vectors
    dim1,n=array1.shape
    dim2,n=array2.shape
    out_arr=np.zeros((dim1*dim2,2))
    c=0
    for a in range(dim1):
        for b in range(dim2):
            out_arr[c,:]=[array1[a],array2[b]]
            c=c+1
    return out_arr
def cartesian_product_coords(array1,array2):
    #gives cartesian product of 2 vectors
    dim1,n1=array1.shape
    dim2,n2=array2.shape
    out_arr=np.zeros((dim1*dim2,n1+n2))
    c=0
    for a in range(dim1):
        for b in range(dim2):
            #print(array1[a,:])
            #print(array2[b,:])
            out_arr[c,:]=np.append(array1[a,:],array2[b,:])
            c=c+1
    return out_arr
def cartesian_product_dofs(array1,array2):
    #gives cartesian product of 2 1d vectors
    dim1=len(array1)
    dim2=len(array2)
    out_arr=np.zeros((dim1*dim2,2))
    c=0
    for a in range(dim1):
        for b in range(dim2):
            out_arr[c,:]=[array1[a],array2[b]]
            c=c+1
    return out_arr


def cartesian_form_to_kroneck_form(indeces, len_dim_2):
    #designed to take list of indexes in the cartesian form
    #translate to single form as kron product would
    #only for cartesian product between 2 spaces
    #expects matrix with n rows, 2 columns
    num_indeces=indeces.shape[0]
    out_arr=np.zeros(num_indeces)
    for n in range(num_indeces):
        out_arr[n] = indeces[n,0]*len_dim_2 + indeces[n,1]
    return out_arr


```

```python
def cartesian_form_to_kroneck_form_3d(cartesian_dofs, len_dof2, len_dof3):
    kronecker_dofs = cartesian_dofs[:, 0] * len_dof2 * len_dof3 + cartesian_dofs[:, 1] * len_dof3 + cartesian_dofs[:, 2]
    return kronecker_dofs


def cartesian_product_dofs_2(array1, array2, array3):
    # da el producto cartesiano de 3 arrays 1D
    dim1 = len(array1)
    dim2 = len(array2)
    dim3 = len(array3)
    out_arr = np.zeros((dim1*dim2*dim3, 3))
    c = 0
    for a in range(dim1):
        for b in range(dim2):
            for d in range(dim3):
                out_arr[c, :] = [array1[a], array2[b], array3[d]]
                c = c + 1
    return out_arr
```

## **Heat equation in 4D**

# 4D espaciales y 1D temporal
$$\frac{\partial u }{\partial t}=(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}+\frac{\partial^2 u}{\partial z^2}+\frac{\partial^2 u}{\partial w^2}),\quad in\quad \Omega 𝗫(0,T]= \Omega_T$$$$, \quad u=u_D \quad on \quad ∂Ω𝗫(0,T]$$$$u=u_0 \quad in\quad t=0$$

```python
# Ecuacion de calor para 4D
start_time = time.time()


alpha = 3          # parametro alpha
beta = 1.2         # parametro beta

num_elem = 8
num_t = 10
mesh1 = dl.UnitSquareMesh(num_elem,num_elem)
mesh2 = dl.UnitSquareMesh(num_elem,num_elem)
mesh3 = dl.UnitIntervalMesh(num_t)

# Creamos las funciones para cada espacio
V1 = dl.FunctionSpace(mesh1, 'P', 1)
V2 = dl.FunctionSpace(mesh2, 'P', 1)
V3 = dl.FunctionSpace(mesh3, 'P', 1)

# Creamos las funciones de trial and test
u1_trial = dl.TrialFunction(V1)
v1_test = dl.TestFunction(V1)
u2_trial = dl.TrialFunction(V2)
v2_test = dl.TestFunction(V2)
u3_trail = dl.TrialFunction(V3)
v3_test = dl.TestFunction(V3)

#Ensamblamos las formas debiles (complejas debido a la regla de la cadena)
stiffness_form11 = dl.inner(dl.grad(u1_trial), dl.grad(v1_test)) * dl.dx
stiffness_form12 = u1_trial*v1_test * dl.dx
stiffness_form13 = u1_trial*v1_test * dl.dx

stiffness_form21 = dl.inner(dl.grad(u2_trial), dl.grad(v2_test)) * dl.dx
stiffness_form22 = u2_trial*v2_test * dl.dx
stiffness_form23 = dl.inner(dl.grad(u3_trail),dl.grad(v3_test)) * dl.dx -u3_trail.dx(0)*v3_test*dl.ds

stiffness_form31 = u3_trail*v3_test * dl.dx
stiffness_form32 = dl.inner(dl.grad(u2_trial), dl.grad(v2_test)) * dl.dx
stiffness_form33 = u3_trail*v3_test * dl.dx

#asignado el lado izquierdo de la ecuacion ( necesitaremos generalizarla mas tarde)
#f = dl.constant(#)

#Ensamblando las matrices stiffness (solo cuando trabajambos con formas lineales)
K11 = dl.assemble(stiffness_form11)
K12 = dl.assemble(stiffness_form12)
K13 = dl.assemble(stiffness_form13)

K21 = dl.assemble(stiffness_form21)
K22 = dl.assemble(stiffness_form22)
K23 = dl.assemble(stiffness_form23)

K31 = dl.assemble(stiffness_form31)
K32 = dl.assemble(stiffness_form32)
K33 = dl.assemble(stiffness_form33)

# Asignado la parte derecha (cuando el termimino de fuerza es 0)

b1 = dl.Function(V1)
b2 = dl.Function(V2)
b3 = dl.Function(V3)


#usamos estas funciones para marcar el contorno (Dirichlet)

def boundary(x, on_boundary):
  return on_boundary

def time_boundary(x, on_boundary):
  if on_boundary and np.abs(x-0.0) < 1e-8:
    return True

  else:
    return False
#esta funcion es usada simplemente para marcar el contorno

u_D1 = dl.Expression('1.0', degree=2)

#estableciendo funciones ficticias
dum1=dl.Function(V1)
dum2=dl.Function(V2)
dum3=dl.Function(V3)

#creamos las condiciones de frontera
bc1= dl.DirichletBC(V1, u_D1, boundary)
bc2= dl.DirichletBC(V2, u_D1, boundary)
bc3= dl.DirichletBC(V3, u_D1, boundary)

#aplicando los vectores a las marcas del contorno
bc1.apply(dum1.vector())
bc2.apply(dum2.vector())
bc3.apply(dum3.vector())

#obteniendo el numero de indice para cada coordenada del limite en el subdominio
boundary_dofs1 = np.where(dum1.vector()==1.0)[0]
boundary_dofs2 = np.where(dum2.vector()==1.0)[0]
boundary_dofs3 = np.where(dum3.vector()==1.0)[0]

#Mostramos los indices del limite en el subdominio
print(boundary_dofs1)
print(boundary_dofs2)
print(boundary_dofs3)

#necesitaremos todas las coordenadas fisicas por si necesitamos hacer caculos

dof_coordinates1 = V1.tabulate_dof_coordinates()
dof_coordinates2 = V2.tabulate_dof_coordinates()
dof_coordinates3 = V3.tabulate_dof_coordinates()

#las coordenadas físicas deben ser solo el producto cartesiano de 2 mapas dof en el mismo orden en que opera el producto kron

global_1 = cartesian_product_coords(dof_coordinates1, dof_coordinates2)
global_dof = cartesian_product_coords(global_1, dof_coordinates3)

#coordenadas del contorno en cada subdominio
boundary_coord1 = dof_coordinates1[boundary_dofs1]
boundary_coord2 = dof_coordinates2[boundary_dofs2]
boundary_coord3 = dof_coordinates3[boundary_dofs3]

#ahora conecte los índices de subdominio al contorno global (convierta esto en función en el futuro)
#tal vez en el futuro, para la velocidad, podamos calcular previamente el número de grados de libertad de los límites
global_boundary_dofs = np.empty(((len(boundary_dofs1))*len(dof_coordinates2)*len(dof_coordinates3) +
                                len(boundary_coord2)*len(dof_coordinates1)*len(dof_coordinates3) +
                                len(boundary_coord3)*len(dof_coordinates1)*len(dof_coordinates2), 3), dtype=int)
ctr = 0
for j in boundary_dofs1:
    global_boundary_dofs[ctr*len(dof_coordinates2)*len(dof_coordinates3):(ctr+1)*len(dof_coordinates2)*len(dof_coordinates3),:] = \
    cartesian_product_dofs_2(np.array([j]), np.arange(dof_coordinates2.shape[0]), np.arange(dof_coordinates3.shape[0]))
    ctr = ctr + 1

last_ind = (ctr) * len(dof_coordinates2) * len(dof_coordinates3)

for j in boundary_dofs2:
    global_boundary_dofs[last_ind:last_ind+len(dof_coordinates1)*len(dof_coordinates3),:] = \
    cartesian_product_dofs_2(np.arange(dof_coordinates1.shape[0]), np.array([j]), np.arange(dof_coordinates3.shape[0]))
    last_ind = last_ind + len(dof_coordinates1) * len(dof_coordinates3)

for j in boundary_dofs3:
    global_boundary_dofs[last_ind:last_ind+len(dof_coordinates1)*len(dof_coordinates2),:] = \
    cartesian_product_dofs_2(np.arange(dof_coordinates1.shape[0]), np.arange(dof_coordinates2.shape[0]), np.array([j]))
    last_ind = last_ind + len(dof_coordinates1) * len(dof_coordinates2)

#cortar y tambien eliminar los contornos duplicados
global_boundary_dofs = np.unique(global_boundary_dofs, axis=0)

#tenemos un producto cartesiano de dof en todo el contorno (esta forma deberia se facil para obtener las coodenadas si las necestamos)
#Ahora necesitamos convertir el global system dof con una funcion de kronecker
global_boundary_dofs = cartesian_form_to_kroneck_form_3d(global_boundary_dofs, len(dof_coordinates2), len(dof_coordinates3))
global_boundary_dofs = global_boundary_dofs.astype(int)



#ensamblamos el sistema global usando kronecker product
K11 = np.array(K11.array())
K12 = np.array(K12.array())
K13 = np.array(K13.array())

K21 = np.array(K21.array())
K22 = np.array(K22.array())
K23 = np.array(K23.array())

K31 = np.array(K31.array())
K32 = np.array(K32.array())
K33 = np.array(K33.array())

#
A = np.kron(np.kron(K11,K22),K33) + np.kron(np.kron(K12,K31),K21) + np.kron(np.kron(K13,K32),K23)



b1 = b1.vector()[:]
b2 = b2.vector()[:]
b3 = b3.vector()[:]

B = np.kron(np.kron(b1,b2),b3)

#necesitamos aplicar bcs usando la informacion de boundary_dofs
# estabeceremos las filas de 1 sobre la frontera en una diagonal
K_bound = np.zeros(A.shape)

for i in global_boundary_dofs:
    K_bound[i,i] = 1

A[global_boundary_dofs,:] = K_bound[global_boundary_dofs,:]


#set the value
#B[global_boundary_dofs] = 1 + (global_dof[global_boundary_dofs,0])**2 + alpha*(global_dof[global_boundary_dofs,1])**2 + alpha*(global_dof[global_boundary_dofs,2])**2 +alpha*(global_dof[global_boundary_dofs,3])**2 + alpha*(global_dof[global_boundary_dofs,4])**2 + beta*global_dof[global_boundary_dofs,2]
B[global_boundary_dofs] = np.sin(np.pi*global_dof[global_boundary_dofs,0])*np.sin(np.pi*global_dof[global_boundary_dofs,1])*np.sin(np.pi*global_dof[global_boundary_dofs,3])*np.sin(np.pi*global_dof[global_boundary_dofs,4])

#solve global system
u=np.zeros(len(B))

#solve(A,u,B)
u=np.linalg.solve(A,B)


end_time = time.time()
time4 = end_time - start_time
Time_execution.append(time4)
print("Tiempo de ejecución:", time4, "segundos")

# Uso de la memoria
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"Uso de memoria: {mem_info.rss / (1024 ** 2):.2f} MB")
```

$$\frac{\partial u }{\partial t}=(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}+\frac{\partial^2 u}{\partial z^2}+\frac{\partial^2 u}{\partial w^2}+\frac{\partial^2 u}{\partial v^2}),\quad in\quad \Omega 𝗫(0,T]= \Omega_T$$$$, \quad u=u_D \quad on \quad ∂Ω𝗫(0,T]$$$$u=u_0 \quad in\quad t=0$$

```python
# Ecuacion de calor para 5D

start_time = time.time()
alpha = 3          # parametro alpha
beta = 1.2         # parametro beta

num_elem = 8
n_elem = 2
num_t = 10
mesh1 = dl.UnitSquareMesh(num_elem,num_elem)
mesh2 = dl.UnitCubeMesh(n_elem,n_elem, n_elem)
mesh3 = dl.UnitIntervalMesh(num_t)

# Creamos las funciones para cada espacio
V1 = dl.FunctionSpace(mesh1, 'P', 1)
V2 = dl.FunctionSpace(mesh2, 'P', 1)
V3 = dl.FunctionSpace(mesh3, 'P', 1)

# Creamos las funciones de trial and test
u1_trial = dl.TrialFunction(V1)
v1_test = dl.TestFunction(V1)
u2_trial = dl.TrialFunction(V2)
v2_test = dl.TestFunction(V2)
u3_trail = dl.TrialFunction(V3)
v3_test = dl.TestFunction(V3)

#Ensamblamos las formas debiles (complejas debido a la regla de la cadena)
stiffness_form11 = dl.inner(dl.grad(u1_trial), dl.grad(v1_test)) * dl.dx
stiffness_form12 = u1_trial*v1_test * dl.dx
stiffness_form13 = u1_trial*v1_test * dl.dx

stiffness_form21 = dl.inner(dl.grad(u2_trial), dl.grad(v2_test)) * dl.dx
stiffness_form22 = u2_trial*v2_test * dl.dx
stiffness_form23 = dl.inner(dl.grad(u3_trail),dl.grad(v3_test)) * dl.dx #-u3_trail.dx(0)*v3_test*dl.ds

stiffness_form31 = u3_trail*v3_test * dl.dx
stiffness_form32 = dl.inner(dl.grad(u2_trial), dl.grad(v2_test)) * dl.dx
stiffness_form33 = u3_trail*v3_test * dl.dx

#asignado el lado izquierdo de la ecuacion ( necesitaremos generalizarla mas tarde)
#f = dl.constant(#)

#Ensamblando las matrices stiffness (solo cuando trabajambos con formas lineales)
K11 = dl.assemble(stiffness_form11)
K12 = dl.assemble(stiffness_form12)
K13 = dl.assemble(stiffness_form13)

K21 = dl.assemble(stiffness_form21)
K22 = dl.assemble(stiffness_form22)
K23 = dl.assemble(stiffness_form23)

K31 = dl.assemble(stiffness_form31)
K32 = dl.assemble(stiffness_form32)
K33 = dl.assemble(stiffness_form33)

# Asignado la parte derecha (cuando el termimino de fuerza es 0)

b1 = dl.Function(V1)
b2 = dl.Function(V2)
b3 = dl.Function(V3)


#usamos estas funciones para marcar el contorno (Dirichlet)

def boundary(x, on_boundary):
  return on_boundary

def time_boundary(x, on_boundary):
  if on_boundary and np.abs(x-0.0) < 1e-8:
    return True

  else:
    return False
#esta funcion es usada simplemente para marcar el contorno

u_D1 = dl.Expression('1.0', degree=2)

#estableciendo funciones ficticias
dum1=dl.Function(V1)
dum2=dl.Function(V2)
dum3=dl.Function(V3)

#creamos las condiciones de frontera
bc1= dl.DirichletBC(V1, u_D1, boundary)
bc2= dl.DirichletBC(V2, u_D1, boundary)
bc3= dl.DirichletBC(V3, u_D1, boundary)

#aplicando los vectores a las marcas del contorno
bc1.apply(dum1.vector())
bc2.apply(dum2.vector())
bc3.apply(dum3.vector())

#obteniendo el numero de indice para cada coordenada del limite en el subdominio
boundary_dofs1 = np.where(dum1.vector()==1.0)[0]
boundary_dofs2 = np.where(dum2.vector()==1.0)[0]
boundary_dofs3 = np.where(dum3.vector()==1.0)[0]

#Mostramos los indices del limite en el subdominio
print(boundary_dofs1)
print(boundary_dofs2)
print(boundary_dofs3)

#necesitaremos todas las coordenadas fisicas por si necesitamos hacer caculos

dof_coordinates1 = V1.tabulate_dof_coordinates()
dof_coordinates2 = V2.tabulate_dof_coordinates()
dof_coordinates3 = V3.tabulate_dof_coordinates()

#las coordenadas físicas deben ser solo el producto cartesiano de 2 mapas dof en el mismo orden en que opera el producto kron

global_1 = cartesian_product_coords(dof_coordinates1, dof_coordinates2)
global_dof = cartesian_product_coords(global_1, dof_coordinates3)

#coordenadas del contorno en cada subdominio
boundary_coord1 = dof_coordinates1[boundary_dofs1]
boundary_coord2 = dof_coordinates2[boundary_dofs2]
boundary_coord3 = dof_coordinates3[boundary_dofs3]

#ahora conecte los índices de subdominio al contorno global (convierta esto en función en el futuro)
#tal vez en el futuro, para la velocidad, podamos calcular previamente el número de grados de libertad de los límites
global_boundary_dofs = np.empty(((len(boundary_dofs1))*len(dof_coordinates2)*len(dof_coordinates3) +
                                len(boundary_coord2)*len(dof_coordinates1)*len(dof_coordinates3) +
                                len(boundary_coord3)*len(dof_coordinates1)*len(dof_coordinates2), 3), dtype=int)
ctr = 0
for j in boundary_dofs1:
    global_boundary_dofs[ctr*len(dof_coordinates2)*len(dof_coordinates3):(ctr+1)*len(dof_coordinates2)*len(dof_coordinates3),:] = \
    cartesian_product_dofs_2(np.array([j]), np.arange(dof_coordinates2.shape[0]), np.arange(dof_coordinates3.shape[0]))
    ctr = ctr + 1

last_ind = (ctr) * len(dof_coordinates2) * len(dof_coordinates3)

for j in boundary_dofs2:
    global_boundary_dofs[last_ind:last_ind+len(dof_coordinates1)*len(dof_coordinates3),:] = \
    cartesian_product_dofs_2(np.arange(dof_coordinates1.shape[0]), np.array([j]), np.arange(dof_coordinates3.shape[0]))
    last_ind = last_ind + len(dof_coordinates1) * len(dof_coordinates3)

for j in boundary_dofs3:
    global_boundary_dofs[last_ind:last_ind+len(dof_coordinates1)*len(dof_coordinates2),:] = \
    cartesian_product_dofs_2(np.arange(dof_coordinates1.shape[0]), np.arange(dof_coordinates2.shape[0]), np.array([j]))
    last_ind = last_ind + len(dof_coordinates1) * len(dof_coordinates2)

#cortar y tambien eliminar los contornos duplicados
global_boundary_dofs = np.unique(global_boundary_dofs, axis=0)

#tenemos un producto cartesiano de dof en todo el contorno (esta forma deberia se facil para obtener las coodenadas si las necestamos)
#Ahora necesitamos convertir el global system dof con una funcion de kronecker
global_boundary_dofs = cartesian_form_to_kroneck_form_3d(global_boundary_dofs, len(dof_coordinates2), len(dof_coordinates3))
global_boundary_dofs = global_boundary_dofs.astype(int)



#ensamblamos el sistema global usando kronecker product
K11 = np.array(K11.array())
K12 = np.array(K12.array())
K13 = np.array(K13.array())

K21 = np.array(K21.array())
K22 = np.array(K22.array())
K23 = np.array(K23.array())

K31 = np.array(K31.array())
K32 = np.array(K32.array())
K33 = np.array(K33.array())

#
A = np.kron(np.kron(K11,K22),K33) + np.kron(np.kron(K12,K31),K21) + np.kron(np.kron(K13,K32),K23)



b1 = b1.vector()[:]
b2 = b2.vector()[:]
b3 = b3.vector()[:]

B = np.kron(np.kron(b1,b2),b3)

#necesitamos aplicar bcs usando la informacion de boundary_dofs
# estabeceremos las filas de 1 sobre la frontera en una diagonal
K_bound = np.zeros(A.shape)

for i in global_boundary_dofs:
    K_bound[i,i] = 1

A[global_boundary_dofs,:] = K_bound[global_boundary_dofs,:]


#B[global_boundary_dofs] = 1 + (global_dof[global_boundary_dofs,0])**2 + \
#alpha*(global_dof[global_boundary_dofs,1])**2 + alpha*(global_dof[global_boundary_dofs,2])**2 + \
#alpha*(global_dof[global_boundary_dofs,3])**2 + alpha*(global_dof[global_boundary_dofs,4])**2 + \
#alpha*(global_dof[global_boundary_dofs,5])**2 + beta*global_dof[global_boundary_dofs,2]
B[global_boundary_dofs] = np.sin(np.pi*global_dof[global_boundary_dofs,0])*np.sin(np.pi*global_dof[global_boundary_dofs,1])*np.sin(np.pi*global_dof[global_boundary_dofs,3])*np.sin(np.pi*global_dof[global_boundary_dofs,4])*np.sin(np.pi*global_dof[global_boundary_dofs,5])


#solve global system
u=np.zeros(len(B))

#solve(A,u,B)
u=np.linalg.solve(A,B)


end_time = time.time()
time5 = end_time - start_time
Time_execution.append(time5)
print("Tiempo de ejecución:", time5, "segundos")

# Uso de la memoria
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"Uso de memoria: {mem_info.rss / (1024 ** 2):.2f} MB")
```
